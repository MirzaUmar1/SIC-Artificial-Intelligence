##**Practice Tasks: Seaborn**

###**Task 1: Basic Histogram**

####**Create a histogram of the 'mpg' values using Seaborn.**

*   Use the sns.histplot() function to create a histogram.
*   Set bins to 30 and choose a color of your choice.
*   Display the plot using plt.show().

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
auto = sns.load_dataset('mpg')
auto.head()
```

```python
sns.histplot(data=auto, x='mpg', bins=30, color='orange')
```

###**Task 2: KDE Plot with Customization**

####**Create a Kernel Density Estimate (KDE) plot for the 'mpg' values.**

*   Use the sns.kdeplot() function to create the KDE plot.
*   Customize the line style to dashed and set the color to 'purple'.
*   Add a title to the plot: "KDE Plot of MPG Values".
*   Display the plot using plt.show().



```python
sns.kdeplot(data=auto, x='mpg', color='purple', linestyle='--')
plt.title('KDE Plot of MPG Values')
plt.show()
```

###**Task 3: Overlaying Histogram and KDE**

####**Overlay a histogram and KDE plot on the same graph.**



*   Create a histogram of the 'mpg' values using sns.histplot() with bins=50 and color it
'orange'.

*   Overlay a KDE plot using sns.kdeplot() with a color of your choice (e.g., 'blue').
*   Add a legend to distinguish between the histogram and KDE.


*   Display the plot using plt.show().



```python
sns.histplot(data=auto, x='mpg', bins=50, color='orange')
sns.kdeplot(data=auto, x='mpg', color='blue')
plt.legend(labels=['Histogram', 'KDE'])
plt.show()
```

###**Task 4: Exploring Other Features of the Dataset**

####**Explore the relationship between 'mpg' and 'horsepower'.**

*   Use sns.scatterplot() to create a scatter plot of 'mpg' vs 'horsepower'.

*   Set the color of the points to 'green' and add a title: "Scatter Plot of MPG vs
Horsepower".
*   Display the plot using plt.show().



```python
sns.scatterplot(data=auto, x='mpg', y='horsepower', color='green')
plt.title('Scatter Plot of MPG vs Horsepower')
plt.show()
```

###**Task 5: Binning and Customizing Histogram**

####**Objective: Create a histogram with customized bins and additional features.**
