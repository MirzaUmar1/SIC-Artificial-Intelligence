##**Can you create a DataFrame named student_data with the following columns: name, age, major, and GPA? Use the following sample data:**

*   name: ['Alice', 'Bob', 'Charlie', 'David']
*   age: [20, 21, 22, 23]
*   major: ['Math', 'Science', 'Arts', 'Engineering']
*   GPA: [3.5, 3.8, 3.2, 3.9]




```python
name = ['Alice', 'Bob', 'Charlie', 'David']
age = [20,21,22,23]
major = ['Math', 'Science', 'Arts', 'Engineering']
GPA = [3.5,3.8,3.2,3.9]
student_data = pd.DataFrame({'name':name,'age':age,'major':major,'GPA':GPA})
student_data
```

##**Select Specific Columns:**

###**From the student_data DataFrame you created, how would you select and print only the name and GPA columns? Write the code to accomplish this.**

```python
student_data[['name','GPA']]
```

###**How can you add a new column called graduated to the student_data DataFrame, indicating whether a student has graduated based on their GPA (assume GPA >= 3.0 means graduated)? Provide the code to add this column and print the updated DataFrame.**

##**Create a DataFrame with Missing Values:**

###**Can you create a DataFrame named course_data with the following columns: course, credits, and enrollment? Use the following data, where some values are missing:**
*   course: ['Math', 'Science', None, 'Engineering']
*   credits: [3, 4, 3, None]
*   enrollment: [30, 25, 20, 15]

```python
data = {'course':['Math', 'Science', None, 'Engineering'],'credits':[3, 4, 3, None],'enrollment':[30, 25, 20, 15]}
course_data = pd.DataFrame(data)
course_data
```

##**Fill Missing Values**

###**How can you fill the missing values in the credits column of the course_data DataFrame with the average credits? Write the code to perform this operation and print the updated DataFrame.**

```python
course_data['credits'].fillna(course_data['credits'].mean(),inplace=True)
course_data
```

##**Group By and Aggregate:**

###**Create a DataFrame named sales_data with the following columns: store, month, and sales. Use the following data:**

*   store: ['A', 'A', 'B', 'B', 'C', 'C']
*   month: ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb']
*   sales: [200, 220, 150, 180, 300, 320]


```python
store = ['A', 'A', 'B', 'B', 'C', 'C']
month = ['Jan', 'Feb', 'Jan', 'Feb', 'Jan', 'Feb']
sales = [200, 220, 150, 180, 300, 320]
sales_data = pd.DataFrame({'store':store,'month':month,'sales':sales})
sales_data
```

###**How would you group the data by store and calculate the total sales for each store? Provide the code to perform this aggregation and print the results.**

```python
sales_data.groupby('store')['sales'].sum()
```
