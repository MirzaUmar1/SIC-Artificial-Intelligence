####You are tasked with further exploring the Iris dataset using Pandas, NumPy, and Seaborn. The goal is to analyze and visualize the relationship between sepal area and other features in the dataset. Perform the following steps:

####**1. Calculate Sepal Area:**

Create a new column in the DataFrame that calculates the sepal area (sepal
length * sepal width).

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['sepal_area'] = iris['sepal_length'] * iris['sepal_width']
iris.head()
```

####**2. Classify Sepal Area:**


Classify the sepal area into three categories:

*   Small: Sepal area less than 15
*   Medium: Sepal area between 15 and 25
*   Large: Sepal area greater than 25



Create a new column that stores these categories




```python
iris['sepal_area_catagory'] = np.where(iris['sepal_area'] < 15, 'Small', np.where(iris['sepal_area'] <= 25, 'Medium', 'Large'))
iris.head()
```

####**3. Group and Summarize Data:**

Group the data by species and calculate the mean sepal area for each species.

```python
mean_sepal_length_by_species = iris.groupby('species')['sepal_length'].mean()
mean_sepal_length_by_species
```
